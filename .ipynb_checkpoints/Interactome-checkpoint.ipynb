```python
# plot degree distribution
# plot centrality ( closeness and betweenesss)
import csv
import networkx as nx
import pandas as pd
from tqdm import tqdm
# make Graph
G = nx.DiGraph()
```

```python
# Read in MMP ineractome from string database
data = pd.read_csv('string_interactions.tsv', sep = '\t')
data.head(4)
```

```python
#
df = data[["#node1", "node2"]]
G = nx.from_pandas_edgelist(df, '#node1', 'node2')
```

```python
from matplotlib.pyplot import figure
figure(figsize=(10, 30))
nx.draw_shell(G, with_labels=True)
```

```python
leaderboard = {}
for x in G.nodes:
 leaderboard[x] = len(G[x])
s = pd.Series(leaderboard, name='connections')
df2 = s.to_frame().sort_values('connections', ascending=False)
df2
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(100,100))
deg_cent = nx.algorithms.degree_centrality(G)
deg_cent
```
